```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(0, 11)
y = np.abs(np.cos(np.multiply(x, np.exp(np.cos(x) + np.log(x + 1)))))
print(y)
plt.plot(x, y, marker='o', linestyle='-', c = 'r')
plt.axhline(np.mean(y), color='b', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
